```python
# ===============================
# Milestone 2: EDA & Feature Engineering
# ===============================

# 1. Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# 2. Load the preprocessed dataset
df = pd.read_csv("C:\\Users\\MEGHAMALA\\Downloads\\EasyVisa_Preprocessed_Final.csv")

# 3. Basic dataset understanding
print("Dataset Shape:", df.shape)
print("\nDataset Info:")
df.info()
print("\nStatistical Summary:")
print(df.describe())

# 4. EDA: Visa Status Distribution
plt.figure(figsize=(6,4))
sns.countplot(x="case_status", data=df)
plt.title("Visa Application Status Distribution")
plt.show()

# 5. EDA: Processing Time Distribution
plt.figure(figsize=(6,4))
sns.histplot(df["processing_time_days"], bins=30, kde=True)
plt.title("Processing Time Distribution")
plt.show()

# 6. EDA: Job Experience vs Processing Time
plt.figure(figsize=(6,4))
sns.boxplot(x="has_job_experience", y="processing_time_days", data=df)
plt.title("Processing Time vs Job Experience")
plt.show()

# 7. Correlation Analysis
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Feature Correlation Heatmap")
plt.show()

# 8. Feature Engineering: High Processing Delay
df["high_processing_delay"] = (
    df["processing_time_days"] > df["processing_time_days"].median()
).astype(int)

# 9. Feature Engineering: High Wage Flag
df["high_wage_flag"] = (
    df["prevailing_wage"] > df["prevailing_wage"].mean()
).astype(int)

# 10. Save updated dataset after Milestone 2
df.to_csv("C:\\Users\\MEGHAMALA\\Downloads\\EasyVisa_Milestone2_Final.csv", index=False)

print("\nMilestone 2 completed successfully!")
print("New features added:")
print("- high_processing_delay")
print("- high_wage_flag")
```
